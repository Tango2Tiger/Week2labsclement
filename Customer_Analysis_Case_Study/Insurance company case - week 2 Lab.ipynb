```python
file1 = pd.read_csv('Data/file1.csv')
file2 = pd.read_csv('Data/file2.csv')
file3 = pd.read_csv('Data/file3.csv')
```

```python
df1=pd.DataFrame(file1)
df2=pd.DataFrame(file2)
df3=pd.DataFrame(file3)


df1.rename(columns={'Customer':'customer','ST':'state',
                    'GENDER':'gender',"Education":"education", 
                    "Customer Lifetime Value":"lifetime val","Income": "income",
                    "Monthly Premium Auto":"monthly premium",
                    "Number of Open Complaints":"open complaints","Policy Type":"policy",
                    "Vehicle Class":"vehicle class","Total Claim Amount":"total claim amount"}, inplace=True )

df2.rename(columns={'Customer':'customer','ST':'state',
                    'GENDER':'gender',"Education":"education", 
                    "Customer Lifetime Value":"lifetime val","Income": "income",
                    "Monthly Premium Auto":"monthly premium",
                    "Number of Open Complaints":"open complaints","Policy Type":"policy",
                    "Vehicle Class":"vehicle class","Total Claim Amount":"total claim amount"}, inplace=True)

df3.rename(columns={'Customer':'customer','State':'state',
                    'Gender':'gender',"Education":"education", 
                    "Customer Lifetime Value":"lifetime val","Income": "income",
                    "Monthly Premium Auto":"monthly premium",
                    "Number of Open Complaints":"open complaints","Policy Type":"policy",
                    "Vehicle Class":"vehicle class","Total Claim Amount":"total claim amount"}, inplace=True )

```

```python
df=pd.concat([df1,df2,df3], axis=0)
df=df.reset_index(drop=True)
```

```python
df=df.drop_duplicates()
df=df.reset_index(drop=True)
df=df.drop([1071])
df=df.reset_index(drop=True)
```

```python
def stateclean(state):
    if state=="Cali":
        return "California"
    elif state=="AZ":
        return "Arizona"
    elif state=="WA":
        return "Washington"
    else:
        return state

df["state"]=df["state"].apply(stateclean)
df
```

```python
def genderclean(gender):
    if gender in ['M', 'Male',"nan"]:
        return 'male'
    elif gender in ['F', 'female',"Femal","nan"]:
        return 'female'
    else:
        return "male"
    

df["gender"]=df["gender"].apply(genderclean)
df
```

```python
def cleanedu(edu):
    if edu=="Bachelors":
        return "bachelor"
    elif edu==np.nan:
        pass
    else:
        edu=str(edu).lower()
        return edu

df["education"]=df["education"].apply(cleanedu)
```

```python
def cleanint(val):
    if str(val).endswith("%"):
        val = int(round(float(val.rstrip(val[-1]))/100,0))
        return val
    else:
        try:
            return int(round(val))
        except ValueError:
            return val

df["lifetime val"] = df["lifetime val"].apply(cleanint)
df
```

```python
df=df.drop("customer",axis=1)
```

```python
def cleancomplaints(comp):
    try:
        comp=int(comp)
        return comp
    except ValueError:
        if type(comp) is float:
            return 10101
        else:
            comp=comp.split("/")
            comp=comp[1]
            return int(comp)
        
        
df["open complaints"]=df["open complaints"].apply(cleancomplaints)
```

```python
df["total claim amount"] = df["total claim amount"].apply(round,ndigits=1)
```

```python
#replaces NaNs with the mean
mean_val=round(df["lifetime val"].mean(skipna=True))

df["lifetime val"]=df["lifetime val"].fillna(mean_val)
df["lifetime val"]=df["lifetime val"].apply(int)
```

```python
#replaces 0s with mean income

def replacezeros(zero):
    if zero==0:
        return np.nan
    else:
        return zero

df["income"]=df["income"].apply(replacezeros)
    
mean_income=round(df["income"].mean(skipna=True),1)
df["income"]=df["income"].replace(np.nan,mean_income)
```

```python
def bucket_state(state):
    if state=="California":
        return "west region"
    elif state=="Oregon":
        return "north west region"
    elif state=="Washington":
        return "east region"
    elif state=="Arizona" or state=="Nevada":
        return "central region"

df["state"]=df["state"].apply(bucket_state)
df
```

```python
def luxcar(car):
    if car=="Luxury Car" or car=="Luxury SUV":
        return "luxury vehicle"
    else:
        return car.lower()

df["vehicle class"]=df["vehicle class"].apply(luxcar)
```

```python
from scipy import stats


def remove_outliers(columns):
    global df
    for column in columns:
        IQR_multi=stats.iqr(column)*1.5
        lower=column.quantile(0.25)-IQR_multi
        upper=column.quantile(0.75)+IQR_multi
        df=df[column>lower]
        df=df[column<upper]
    return df
```

```python
num_columns=[df["lifetime val"],df["income"],df["monthly premium"],df["total claim amount"]]

df=remove_outliers(num_columns)
```
